```python
df['Total_Alloying_Elements'] = df[['Ni','Si','Co','Cr','Mg']].sum(axis=1)
```

```python
df['thermal_Stress'] = df['Tss'] * df['tss']
```

```python
df['Ni_Si_Ratio'] = df['Ni'] / (df['Si'] + 1e-6)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df['reduction'] = sc.fit_transform(df[['reduction']])
df['hardness'] = sc.fit_transform(df[['hardness']])
df['Tag'] = sc.fit_transform(df[['Tag']])
#df['Tss'] = sc.fit_transform(df[['Tss']])
df['tag'] = sc.fit_transform(df[['tag']])
#df['tss'] = sc.fit_transform(df[['tss']])
df['Total_Alloying_Elements'] = sc.fit_transform(df[['Total_Alloying_Elements']])
df['thermal_Stress'] = sc.fit_transform(df[['thermal_Stress']])
df['Ni_Si_Ratio'] = sc.fit_transform(df[['Ni_Si_Ratio']])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Aging'] = le.fit_transform(df['Aging'])
df['Secondary thermo-mechanical process'] = le.fit_transform(df['Secondary thermo-mechanical process'])

```

```python
y = df['electrical']
x = df.drop(['ID','Alloy formula','Alloy class', 'yield' ,	'tensile','tss','Tss',	'electrical','Ni','Si','Co','Cr','Mg'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
rf.score(x_test,y_test)
```
